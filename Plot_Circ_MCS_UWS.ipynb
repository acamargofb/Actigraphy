# Plot of Circadian results

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.io
import pandas as pd
import numpy as np
```

## Loading the data

```python
mat = scipy.io.loadmat('/home/aldo/Documents/Projects/Avtivemeter/Data/MCS_UWS_LombScargle.mat')
```

```python
mcs = mat['table_results'][0][0][0]
uws = mat['table_results'][0][0][1]
```

```python
mcsp = {'Circadian':mcs[:,0],'Power':mcs[:,1]}
uwsp = {'Circadian':uws[:,0],'Power':uws[:,1]}
dfmcs=pd.DataFrame(mcsp)
dfuws=pd.DataFrame(uwsp)
```

```python
dfmcs['Diagnosis'] = None
```

```python
MCS = np.array('MCS')
mcs_list = np.matlib.repmat(MCS,len(mcs),1)
UWS = np.array('UWS')
uws_list = np.matlib.repmat(UWS,len(uws),1)
```

```python
dfmcs["Diagnosis"] = mcs_list
dfuws["Diagnosis"] = uws_list
```

```python
frames = [dfmcs, dfuws]
result_df = pd.concat(frames)
```

## Error bar plot of MCS and UWS

```python
sns.set_style('darkgrid')
sns.catplot(x="Diagnosis", y="Power", kind="bar", data=result_df)
plt.savefig('/home/aldo/Desktop/errorbar_plot_circadian_mcs_uws.jpg')
```

## Loading the data

```python
mat = scipy.io.loadmat('/home/aldo/Documents/Projects/Avtivemeter/Data/MCS_UWS_all_LombScargle_circadian.mat')
```

```python
mcs = mat['table_results'][0][0][0]
uws = mat['table_results'][0][0][1]
emcs = mat['table_results'][0][0][2]
lis = mat['table_results'][0][0][3]
control = mat['table_results'][0][0][4]
```

```python
mcsp = {'Circadian':mcs[:,0],'Power':mcs[:,1]}
uwsp = {'Circadian':uws[:,0],'Power':uws[:,1]}
emcsp = {'Circadian':emcs[:,0],'Power':emcs[:,1]}
lisp = {'Circadian':lis[:,0],'Power':lis[:,1]}
controlp = {'Circadian':control[:,0],'Power':control[:,1]}
dfmcs=pd.DataFrame(mcsp)
dfuws=pd.DataFrame(uwsp)
dfemcs=pd.DataFrame(emcsp)
dflis=pd.DataFrame(lisp)
dfcontrol=pd.DataFrame(controlp)
```

```python
MCS = np.array('MCS')
mcs_list = np.matlib.repmat(MCS,len(mcs),1)
UWS = np.array('UWS')
uws_list = np.matlib.repmat(UWS,len(uws),1)
EMCS = np.array('EMCS')
emcs_list = np.matlib.repmat(EMCS,len(emcs),1)
LIS = np.array('LIS')
lis_list = np.matlib.repmat(LIS,len(lis),1)
CONTROL = np.array('CONTROL')
control_list = np.matlib.repmat(CONTROL,len(control),1)
```

```python
dfmcs["Diagnosis"] = mcs_list
dfuws["Diagnosis"] = uws_list
dfemcs["Diagnosis"] = emcs_list
dflis["Diagnosis"] = lis_list
dfcontrol["Diagnosis"] = control_list
```

```python
frames = [dfmcs, dfuws, dfemcs, dflis, dfcontrol]
result_df = pd.concat(frames)
```

## Error bar plot of MCS, UWS, EMCS, LIS and Control

```python
sns.set_style('darkgrid')
sns.catplot(x="Diagnosis", y="Power", kind="bar", data=result_df)
plt.savefig('/home/aldo/Desktop/errorbar_plot_circadian_all.jpg')
```

```python
sns.set_style('darkgrid')
sns.catplot(x="Diagnosis", y="Power", kind="box", data=result_df)
#plt.savefig('/home/aldo/Desktop/bar_plot_circadian_all.jpg')
```
